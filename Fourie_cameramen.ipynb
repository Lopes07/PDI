```python
import numpy as np
import matplotlib.pyplot as plt
from skimage.io import imread, imshow
from skimage.color import rgb2gray
```

```python
# carrega imagem
Image = imread('cameraman.tif')
# Converte para escala de cinza se for uma imagem colorida RGB
if len(Image.shape) > 2:
    Image = rgb2gray(Image)
```

```python
plt.imshow(Image, cmap='gray')
plt.title('Imagem em tons de cinza');
```

```python
Image_fft = np.fft.fftshift(np.fft.fft2(Image))
plt.imshow(np.log(abs(Image_fft)), cmap='gray')
plt.title('Fourier Transform Representation of the Cargo Ship');
```

```python
Image_fft[:125, 129:132] = 1
Image_fft[-125:,129:132] = 1

plt.imshow(np.log(abs(Image_fft)), cmap='gray')
plt.title('Representação modificada da Transformada de Fourier da imagem');
```

Para remover as linhas brancas da imagem, podemos suprimir a linha vertical, substituindo-a por um valor pequeno (nesse caso, o valor é 1). Depois disso, podemos transformar a imagem da Representação da Transformada de Fourier modificada de volta para o domínio espacial.

```python
fig, ax = plt.subplots(1,2,figsize=(15,15))
ax[0].imshow(Image, cmap='gray')
ax[0].set_title('Imagem em tons de cinza')
ax[1].imshow(abs(np.fft.ifft2(Image_fft)), cmap='gray')
ax[1].set_title('Imagem transformada');
```

```python
Image_fft[129:132, :125] = 1
Image_fft[129:132, -125:] = 1

plt.imshow(np.log(abs(Image_fft)), cmap='gray')
plt.title('Representação modificada da Transformada de Fourier da imagem');
```

```python
fig, ax = plt.subplots(1,2,figsize=(15,15))
ax[0].imshow(Image, cmap='gray')
ax[0].set_title('Imagem em tons de cinza')
ax[1].imshow(abs(np.fft.ifft2(Image_fft)), cmap='gray')
ax[1].set_title('Imagem transformada');
```
