
Ler os m√≥dulos.

```python
from PIL import Image
from PIL.ExifTags import TAGS
```

Abrindo o arquivo pelo modulo Image.

```python
image_b = Image.open('images/backyard.png')
```

Extrair os metadados da imagem

```python
info_dict = {
    "Filename": image_b.filename,
    "Image Size": image_b.size,
    "Image Height": image_b.height,
    "Image Width": image_b.width,
    "Image Format": image_b.format,
    "Image Mode": image_b.mode,
    "Image is Animated": getattr(image_b, "is_animated", False),
    "Frames in Image": getattr(image_b, "n_frames", 1)
}
```

```python
for label,value in info_dict.items():
    print(f"{label:25}: {value}")
```

Abrindo o arquivo pelo modulo Image.

```python
image_g = Image.open('images/galaxy.tif')
```

Extrair os metadados da imagem

```python
exifdata = image_g.getexif()
```

Todos os dados contidos no campo de dados de EXIF

```python
for tag_id in exifdata:
    # tag name em vez de tag id  ileg√≠vel para humanos
    tag = TAGS.get(tag_id, tag_id)
    data = exifdata.get(tag_id)
    # decodificar bytes
    if isinstance(data, bytes):
        data = data.decode()
    print(f"{tag:25}: {data}")
```
