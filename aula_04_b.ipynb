```python
# Importando os modulos
import matplotlib.pyplot as plt
from skimage import io, filters
```

```python
# Carregar a imagem
Image=io.imread('bergen.jpg')
```

```python
# Aplicar limiarização (thresholding)
# Usando o método de Otsu (alternativa ao valor fixo 0.1)
T = Image > filters.threshold_otsu(Image)
```

```python
# Criar figura com dois subplots
plt.figure(figsize=(12,5))

# Imagem original
plt.subplot(1,3,1)
plt.imshow(Image, cmap='gray')
plt.title('Imagem Original')
plt.axis('off')

# Imagem limiarizada
plt.subplot(1,3,2)
plt.imshow(T, cmap='gray')
plt.title('Imagem Limiarizada')
plt.axis('off')



```

```python
# Mostrar histograma para comparação
plt.subplot(1,3,3)
plt.hist(Image.ravel(), 256)
plt.title('Histograma')
plt.xlabel('Intensidade')
plt.ylabel('Contagem')

# Mostrar a figura
plt.tight_layout()
plt.show()
```
