```python
# Importando os modulos
import numpy as np
import matplotlib.pyplot as plt
from skimage import io, data, filters
from scipy import ndimage
```

```python
# Carregar a imagem
Image = io.imread('rice.png')
```

```python
#Aplicar o filtro
#Em Python, usamos scipy.ndimage ou filtros do scikit-image
Im = ndimage.uniform_filter(Image, size=15)
```

```python
# Subtração da imagem média + constante
It = Image - (Im + 20)

# Limiarização com threshold
Ibw = It<120
#Ibw = It < filters.threshold_otsu(It)
# Criar figura
plt.figure(figsize=(10,5))

# Imagem original
plt.subplot(1,2,1)
plt.imshow(Image, cmap='gray')
plt.title('Imagem Original')
plt.axis('off')

# Imagem binarizada
plt.subplot(1,2,2)
plt.imshow(Ibw, cmap='gray')
plt.title('Imagem Binarizada')
plt.axis('off')

plt.tight_layout()
plt.show()
```
