```python
import numpy as np
import matplotlib.pyplot as plt
from skimage import io, util, filters
from scipy.ndimage import rank_filter

# Carregar imagem
I = io.imread('eight.tif')
I = I.astype(float) / 255.0  # Normalizar para [0, 1]

# Adicionar ruídos
Isp = util.random_noise(I, mode='s&p', amount=0.03)
Ig = util.random_noise(I, mode='gaussian', var=0.02)

# Criar kernel 5x5
kernel = np.ones((5, 5))

# Aplicar filtro de ordem (rank filter)
Im = rank_filter(I, rank=24, size=5)
Isp_m = rank_filter(Isp, rank=24, size=5)
Ig_m = rank_filter(Ig, rank=24, size=5)
```

```python
# Plotar imagens
plt.figure(figsize=(15,5))

plt.subplot(1,3,1)
plt.imshow(Im, cmap='gray')
plt.title('Imagem Original Filtrada')
plt.axis('off')

plt.subplot(1,3,2)
plt.imshow(Isp_m, cmap='gray')
plt.title('Ruído Sal e Pimenta Filtrado')
plt.axis('off')

plt.subplot(1,3,3)
plt.imshow(Ig_m, cmap='gray')
plt.title('Ruído Gaussiano Filtrado')
plt.axis('off')

plt.tight_layout()
plt.show()
```
