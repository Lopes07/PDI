```python
# Histogram equalization

import matplotlib.pyplot as plt
import numpy as np
from skimage import io, img_as_ubyte
from skimage.exposure import equalize_hist

# Ler a imagem
I = io.imread('pout.tif')

# Converter a imagem para uint8 (necessário para equalize_hist)
I = img_as_ubyte(I)

# Equalização de histograma
Ieq = equalize_hist(I)

# Criar a figura com subplots
fig, axes = plt.subplots(2, 2, figsize=(10, 10))

# Mostrar as imagens
axes[0, 0].imshow(I, cmap='gray')
axes[0, 0].set_title('Imagem Original')
axes[0, 0].axis('off')

axes[0, 1].imshow(Ieq, cmap='gray')
axes[0, 1].set_title('Imagem Equalizada')
axes[0, 1].axis('off')

# Mostrar os histogramas
axes[1, 0].hist(I.ravel(), bins=256, range=(0, 256))  # Histograma da imagem original
axes[1, 0].set_title('Histograma Original')

axes[1, 1].hist((Ieq * 255).ravel(), bins=256, range=(0, 256))  # Histograma da imagem equalizada
axes[1, 1].set_title('Histograma Equalizado')

# Ajustar layout e mostrar o plot
plt.tight_layout()
plt.show()
```

O que está acontecendo?

1. A imagem 'pout.tif' é lida."
2. A imagem é convertida para uint8, pois a função equalize_hist opera com esse tipo."
3. A equalização de histograma é aplicada à imagem. Esta técnica redistribui as intensidades de pixel para criar um histograma mais uniforme, aumentando o contraste."
4. As imagens original e equalizada são exibidas."
5. Os histogramas das imagens original e equalizada são exibidos.
 - O histograma da imagem original mostra a distribuição original das intensidades."
- O histograma da imagem equalizada mostra uma distribuição mais uniforme, indicando o aumento do contraste."
